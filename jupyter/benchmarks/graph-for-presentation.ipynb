```python
import pandas
import numpy
import matplotlib.pyplot as plt

dates = [
    "01/12/2021",
    "01/04/2022",
    "01/08/2022",
    "08/12/2023",
    "10/12/2023",
    "13/12/2023",
]
names = [
    "cpu_v1.0",
    "cpu_v2.0",
    "cpu_v3.0",
    "gpu_v1.0",
    "gpu_v2.0",
    "gpu_v3.0",
    "gpu_v3.1",
    "gpu_v3.2",
]

cpu_time = numpy.array(
    [4_860_000, 6_740_000, 3_070_000, 155_560, 170_010, 50_366], dtype=float
)
hist_time = numpy.array([6_180_000, 324_000, 47_000])
x1 = range(cpu_time.shape[0])
x2 = range(hist_time.shape[0])
fig, ax = plt.subplots(1, 1)
ax.bar(x1, cpu_time, alpha=0.7, color="blue")
ax.set_yscale("log")
ax.set_ylabel("Time [\u03bcs]")
ax.set_xticks(x1)
ax.set_xticklabels(names[:-2], rotation=30)
ax.set_title("Total execution time")
fig.savefig("simulation_bench.pdf", transparent=True)

fig, ax = plt.subplots(1, 1)
ax.bar(x2, hist_time, alpha=0.7, color="blue")
ax.set_yscale("log")
ax.set_ylabel("Time [\u03bcs]")
ax.set_xticks(x2)
ax.set_xticklabels(
    [
        "cpu_v3.0",
        "gpu_v3.1",
        "gpu_v3.2",
    ],
    rotation=30,
)
ax.set_title("Total execution time")
fig.savefig("histogram_bench.pdf", transparent=True)
```
